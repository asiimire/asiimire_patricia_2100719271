```python
# Step 1: Import necessary libraries


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


print(np.__version__)
print(pd.__version__)
```

```python
# Load dataset

df = pd.read_csv("niceguys.csv")

print(df.head(5))
```

```python
# data cleaning and pre-processsing
# Check for measing values

print(df.isnull().sum())
```

```python
# # fill measing values
# df['living_allowance'] = df['living_allowance'].fillna(df['living_allowance'].median())

# print(df.head(5))


# Assuming you've loaded your data into a DataFrame named df

# Identify non-numeric values (assuming some cleaning is possible)
def is_numeric(x):
  try:
    return pd.to_numeric(x.replace(',', '').replace('Sh.', ''))  # Clean and convert
  except:
    return None

df['living_allowance_numeric'] = df['living_allowance'].apply(is_numeric)

# Fill missing values (including non-numeric) with a placeholder
df['living_allowance_numeric'].fillna(-11414, inplace=True)  # Replace with -11414 (or any placeholder)

# Now you can calculate the median of the numeric column
median_value = df['living_allowance_numeric'].median()
print(median_value)
```

```python
# step 4: EDA exploratory data analysis
# Distribution of students with age and whether they have gf/bf
# Gender who have bf/bf
# Do you eat age, gender, gf or bf
# Living allowance and gf or bf
# Those with gf or bf if they have siblings
# Career and bf or gf
# Correlation heatMap,correlation matrix and ploting heatmap
# Decision making, insight of the data, data analysis conclusion
# 1; Age and gf or bf
# 2: and gf or bf
```

```python
# 1 Distribution of students with age and whether they have gf/bf


# Convert the 'do_you_have_aboyfriend/girlfriend' to a more readable format
df['do_you_have_aboyfriend/girlfriend'] = df['do_you_have_aboyfriend/girlfriend'].replace({1: 'Yes', 0: 'No'})

# Ensure 'age' column does not contain NaN values
df['age'] = df['age'].fillna(df['age'].median())

# Plot the distribution of age based on relationship status
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='age', hue='do_you_have_aboyfriend/girlfriend', multiple='dodge', kde=True)
plt.title('Distribution of Age Based on Relationship Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Have Boyfriend/Girlfriend',labels=['No', 'Yes'])
plt.show()

```

```python
# 1 Distribution of studennts with gender and whether they have gf/bf


# Convert the 'do_you_have_aboyfriend/girlfriend' to a more readable format
df['do_you_have_aboyfriend/girlfriend'] = df['do_you_have_aboyfriend/girlfriend'].replace({1: 'Yes', 0: 'No'})

# Plot the distribution of age based on relationship status
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='sex', hue='do_you_have_aboyfriend/girlfriend', multiple='dodge', kde=True)
plt.title('Distribution of Age Based on Relationship Status')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Have Boyfriend/Girlfriend',labels=['No', 'Yes'])
plt.show()

```

```python
# Living allowance and gf or bf

# 1 Distribution of studennts with gender and whether they have gf/bf


# Convert the 'do_you_have_aboyfriend/girlfriend' to a more readable format
df['do_you_have_aboyfriend/girlfriend'] = df['do_you_have_aboyfriend/girlfriend'].replace({1: 'Yes', 0: 'No'})
df['living_allowance'] = df['living_allowance'].fillna(df['living_allowance'].median())

# Plot the distribution of age based on relationship status
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='living_allowance', hue='do_you_have_aboyfriend/girlfriend', multiple='dodge', kde=True)
plt.title('Distribution of Age Based on Relationship Status')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Have Boyfriend/Girlfriend',labels=['No', 'Yes'])
plt.show()

```
