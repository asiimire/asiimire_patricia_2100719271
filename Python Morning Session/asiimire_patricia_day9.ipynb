```python
# Exploratory data analysis

import pandas as pd

import numpy as np
import seaborn as sns

df = pd.read_csv("crop.csv")

print(df)
```

```python
print(df.isnull().sum())
```

```python
print(df['temperature'].value_counts())
```

```python
# Temperature distribution

sns.histplot(df['temperature'], color='red', kde=True)
plt.title("Temperature distribution")
plt.show()
```

```python
# PH distribution

sns.histplot(df['ph'], color='blue', kde=True)
plt.title("PH distribution")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("crop.csv")

sns.boxplot(data=df, x = 'P', y = 'label')
plt.title("Phosphorus level")

plt.show()
```

```python
sns.countplot(y='label', data=df)
plt.title("PH distribution")
plt.show()
```

```python
# Visualising correlation
# Vidualising correlation
c = df.label.astype('category')
targets = dict(enumerate(c.cat.categories))
df['target'] = c.cat.codes

y = df.target
x = df[['K','N','P','temperature', 'humidity','ph', 'rainfall']]

sns.heatmap(x.corr(), annot=True)
plt.title("Correlation")
plt.show()
```
