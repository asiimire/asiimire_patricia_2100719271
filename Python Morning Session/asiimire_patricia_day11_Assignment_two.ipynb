```python
import pandas as pd
%pip install matplotlib
import matplotlib.pyplot as plt
%pip install seaborn
import seaborn as sns
```

```python
# Load the dataset
data=pd.read_csv("heart.csv")
print(data)
```

```python
# Describing the data
description = data.describe()
print("Statistical summary of the heart.csv dataset:")
print(description)
```

```python
# Checking for null values in the dataset
null_counts = data.isnull().sum()

# Printing the column names along with their respective null value counts
print("Null values in each column:\n")
for column, count in null_counts.items():
    print(f"{column}: {count}")

# import seaborn as sns
# import matplotlib.pyplot as plt
# # Visualizing null values using a heatmap
# plt.figure(figsize=(10, 6))
# sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
# plt.title('Null Values in Dataset')
# plt.show()
```

```python
# Convert the input data to a DataFrame if it's not already
data = pd.DataFrame(data)

# Check for duplicate rows in the DataFrame
duplicates = data.duplicated(keep=False)

# Print the duplicate rows
print("Duplicate rows in the DataFrame:")
print(data[duplicates])

# Count the number of duplicates and unique rows
duplicate_count = duplicates.sum()
unique_count = len(data) - duplicate_count

# Create a bar plot to visualize the count of duplicates and unique rows
plt.figure(figsize=(8, 6))
plt.bar(['Duplicates', 'Unique'], [duplicate_count, unique_count], color=['red', 'blue'])
plt.xlabel('Row Type')
plt.ylabel('Count')
plt.title('Count of Duplicate and Unique Rows')
plt.show()

# the red bar (Duplicates) is significantly high,
# it indicates that there are many duplicate entries in the dataset,
# which might need to be addressed to ensure the quality and integrity of the
# data analysis.
# the blue bar (Unique) is significantly higher than the red bar,
# it suggests that the majority of the dataset consists of unique rows,
# indicating fewer issues with duplicate data.
```

```python
# Displaying original and cleaned data side by side
print("Original Data:\n", data)
cleaned_data = data.drop_duplicates()
print("\nData After Removing Duplicates:\n", cleaned_data)

# Creating a bar plot to compare counts
fig, ax = plt.subplots(figsize=(8, 6))

# Calculate counts
original_counts = [len(data), len(data) - len(cleaned_data)]
cleaned_counts = [len(cleaned_data), 0]

bar_width = 0.35
index = range(2)

# Plotting bars
rects1 = ax.bar(index, original_counts, bar_width, label='Original Data')
rects2 = ax.bar([i + bar_width for i in index], cleaned_counts, bar_width, label='Cleaned Data')

# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Data Type')
ax.set_ylabel('Count')
ax.set_title('Comparison of Original and Cleaned Data')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(['Original', 'Cleaned'])
ax.legend()
```

```python
#Exploratory data analysis
#mean ,mode,and median

# Calculate mean, mode, and median
mean_values = data.mean()
mode_values = data.mode().iloc[0]  # Mode for the entire DataFrame
median_values = data.median()

# Print the results
print("Mean:\n", mean_values)
print("\nMode:\n", mode_values)
print("\nMedian:\n", median_values)


```

```python
#Range.
print("Range:\n", data.max() - data.min())
```

```python
#Variance
print("Variance:\n", data.var())
```

```python
# Calculate the standard deviation for each column
std_deviation = data.std()

# Print the standard deviation
print("Standard Deviation:\n", std_deviation)
```

```python
# Defining a function to calculate IQR
def calculate_iqr(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    return q3 - q1

# Calculating and printing the interquartile range using the custom function
iqr = calculate_iqr(data)
print("Interquartile Range:\n", iqr)
```

```python
#Visualization
#Histogram

# Histogram for Age
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], kde=True, bins=10)
plt.title('Age Distribution')
plt.xlabel('age')
plt.ylabel('sex')
plt.show()
```

```python
#Bar graph
plt.figure(figsize=(6, 6))
sns.countplot(x='sex', data=data, palette='viridis')
plt.title('Distribution of Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()
```

```python
#Heatmap
plt.figure(figsize=(8, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python

# Scatter plot examining the relationship between age and cholesterol
plt.figure(figsize=(6, 6))
sns.scatterplot(x='age', y='chol', data=data)
plt.title('Scatter Plot of Age vs Chol')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.show()
```
